Python libraries

```python
import pandas as pd
import numpy as np
```

Grab all pixel values, reduce the value range from [0,255] to [0,1].

```python
df = pd.read_csv("./files/fashion-mnist_train.csv")
df.iloc[:,1:] = df.iloc[:,1:].astype(np.float64)
num_rows,num_cols = df.shape

df.iloc[:, 1:] = df.iloc[:, 1:].astype(np.float64) / 255.0

df.to_csv("./files/fashion-mnist_train_normalized.csv",index=False)
```
